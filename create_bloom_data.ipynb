# Add Groups for training


```python
Uids = data_df.Uid.unique()
UidDict = dict(zip(Uids, range(len(Uids))))
```

```python
data_df['Uid'] = data_df['Uid'].apply(lambda x: UidDict[x])
```

```python
data_df['group'] = data_df['ProductID'].astype(str) + '_' + data_df['Uid'].astype(str)
```

```python
groups = data_df.group.unique()
GroupDict = dict(zip(groups, range(len(groups))))
data_df['group'] = data_df['group'].apply(lambda x: GroupDict[x])
```

```python
data_df['event'] = data_df['event'].str.replace("pdpUniqueViewImage_","", regex=False)  
```

```python
data_df.head()
```

# Get converted groups and relabel all viewed images

```python
converts = data_df[data_df['conversion'] == 1]
```

```python
conversionDF = data_df[data_df['group'].isin(converts['group'])].reset_index(drop=True)
```

```python
conversionDF['conversion'] = 1
```

# Create User DF with User features and add non-viewed images to groups


```python
UserDF = data_df.groupby(['Uid']).first()
UserDF = UserDF.iloc[:, 117:-1]
UserDF = UserDF.reset_index()
UserDF.head()
```

```python
ImageDF = data_df.groupby(['ImageID']).first()
ImageDF = ImageDF.drop(ImageDF.iloc[:, 0:5], axis=1)
ImageDF = ImageDF.drop(ImageDF.iloc[:, 1:17], axis=1)
ImageDF = ImageDF.drop(ImageDF.iloc[:, 96:], axis=1).reset_index(drop=False)
ImageDF['conversion'] = 0
ImageDF.head()
```

```python
convertDF = conversionDF[['Uid', 'ProductID', 'group']]
convertDF = convertDF.groupby('group').first().reset_index(drop=False)
```

```python
convertDF.head()
```

```python
salesDF = convertDF.merge(ImageDF, how='left', left_on='ProductID', right_on='ProductID')
salesDF.head()
```

```python
convertDF = conversionDF[['ProductID', 'ImageID', 'conversion', 'group']].drop_duplicates()
convertDF = convertDF.set_index(['group', 'ProductID', 'ImageID'])
```

```python
salesDF = salesDF.set_index(['group', 'ProductID', 'ImageID'])
salesDF.update(convertDF)
salesDF = salesDF.reset_index(drop=False)
salesDF['conversion'].value_counts()
```

```python
salesDF.head()
```

# Add conversion rate for each image

```python
countDF = pd.DataFrame(data_df['ImageID'].value_counts()).reset_index()
countDF.columns = ['ImageID', 'counts']
countDF['rate'] = float(0)
countDF.head()
```

```python
for idx, row in countDF.iterrows():
    rate = len(salesDF[(salesDF['ImageID'] == row['ImageID']) & (salesDF['conversion'] == 1)])/row['counts']
    countDF.at[idx, 'rate'] = rate
countDF.head()
```

```python
countDF = countDF.drop('counts', axis=1)
salesDF = salesDF.merge(countDF, how='left', left_on='ImageID', right_on='ImageID')
```

```python
salesDF.head()
```

```python
fullSalesDF = salesDF.merge(UserDF, left_on="Uid", right_on='Uid')
```
